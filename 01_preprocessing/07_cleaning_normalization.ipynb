```python
incorrect_text = """
To jest teskt z kilkoma literuwkami i bledami ortgrafcznymi.
Będziemy mogli sprubowacc jak metoda oparta o odleglosc
pozwoli nam je naprawić.
"""
```

```python
#tokenization
incorrect_text_words = nltk.word_tokenize(incorrect_text)
incorrect_text_words
```

```python
import textdistance
```

```python
def find_closest_word(word):
    if word.lower() in [w.lower() for w in pl_words]:
        return word
    else:
        distances = [textdistance.damerau_levenshtein(word.lower(), w.lower()) for w in pl_words]
        max_similarity = max(distances)

        return pl_words[distances.index(max_similarity)]

correct_text_words = (' ').join(list(map(find_closest_word, incorrect_text_words)))
correct_text_words
```

```python
import unidecode
```

```python
def first_last_char(w1, w2):
    w1 = unidecode.unidecode(w1)
    w2 = unidecode.unidecode(w2)

    return w1[0] == w2[0] and w1[-1] == w2[-1]

def find_closest_word(word, max_diff=2):
    if word.lower() in [w.lower() for w in pl_words]:
        return word
    else:
        filtered_words = filter(lambda w: abs(len(w) - len(word)) <= max_diff, pl_words)
        filtered_words = sorted(filter(lambda w: first_last_char(word, w), filtered_words))
        distances = map(lambda w: textdistance.damerau_levenshtein(w.lower(), word.lower()), filtered_words)

        word_distance = dict(zip(filtered_words, distances))
        try:
            return min(word_distance, key=word_distance.get)
        except ValueError:
            return word
```

```python
%%time
closest_word = find_closest_word("literuwka")
print(closest_word)
```
