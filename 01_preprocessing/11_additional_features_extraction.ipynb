```python
import nltk
nltk.download('twitter_samples')
nltk.download('punkt')
import json
import pandas as pd
```

```python
file_ids = nltk.corpus.twitter_samples.fileids()
file_ids
```

```python
docs = nltk.corpus.twitter_samples.docs(file_ids[0])
print(json.dumps(docs[5], indent=4, sort_keys=True))
```

```python
df = pd.concat(
    [pd.DataFrame(pd.json_normalize(nltk.corpus.twitter_samples.docs(file))) for file in file_ids]
)
df.head()
```

```python
df.sample(n=5).T
```

```python
#most interesting columns
columns = [
    # Tweet related information:
    "text", "retweet_count", "created_at", "favorite_count",
    "user.screen_name", "user.listed_count", "user.utc_offset",
    "user.favourites_count", "user.followers_count",
    "user.created_at", "user.description",
    # Quoted tweet related data, the same columns as above:
    "quoted_status.text", "quoted_status.retweet_count",
    "quoted_status.created_at", "quoted_status.favorite_count",
    "quoted_status.user.screen_name", "quoted_status.user.listed_count",
    "quoted_status.user.favourites_count", "quoted_status.user.followers_count"
]

df[columns].sample(n=5)
```

```python
df[columns].describe(include="all").T
```

```python
filtered = df[columns].drop_duplicates()
```

```python
datetime_columns = ['created_at', 'user.created_at', 'quoted_status.created_at']
for dt_column in datetime_columns:
    filtered[dt_column] = pd.to_datetime(filtered[dt_column])

filtered.sample(n=5)
```

```python
local_time_created_at = filtered['created_at'] + pd.to_timedelta(filtered['user.utc_offset'], unit="seconds")
filtered['created_at_hour'] = local_time_created_at.fillna(filtered['created_at']).dt.hour
filtered['created_at_hour']
```

```python
response_time_hours = (filtered['quoted_status.created_at'] - filtered['created_at']).dt.total_seconds() / 3600
filtered['response_time_hours'] = response_time_hours.fillna(-1)
filtered['response_time_hours']
```

```python
filtered['user_exists_for_days'] = (filtered['created_at'] - filtered['user.created_at']).dt.days
filtered['user_exists_for_days']
```

```python
filtered.drop(columns=['created_at', 'user.created_at', 'user.utc_offset', 'quoted_status.created_at'], inplace=True)
filtered.sample(n=5).T
```

```python
text_columns = ['text', 'user.description', 'quoted_status.text']
categorical_columns = ['user.screen_name', 'quoted_status.user.screen_name']
numerical_columns = filtered.drop(columns=text_columns).drop(columns=categorical_columns).columns.tolist()
```

```python
filtered.head()
```

```python
import nltk
import re
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub('(\s)#', '\\1', text)
    text = re.sub('@([^\s]+)', '<USER_MENTION>', text)
    text = re.sub('(https?://[^\s]+)', '<EXTERNAL_LINK>', text)
    return text
```

```python
preprocess_text(
    'Hey @user, this is #one of the #tests. ' \
    'Please visit: https://www.google.com for more details!'
)
```

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('english')
```

```python
def stemming(text):
    return [stemmer.stem(word) for word in nltk.word_tokenize(text)]
```

```python
stemming('This is just one of the examples')
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#transofrmers = [(name, vectorizer, id) * 3]
text_columns_preprocessing = ColumnTransformer(transformers=[
    (f"tfidf_{i}", TfidfVectorizer(preprocessor=preprocess_text,
                                   tokenizer=stemming,
                                   ngram_range=(1, 2),
                                   decode_error="ignore",
                                   min_df=2, max_df=0.9), i)
    for i, _ in enumerate(text_columns)
])
```

```python
text_pipeline = Pipeline(steps=[
    ("imputer_nan", SimpleImputer(strategy="constant")),
    ("imputer_none", SimpleImputer(missing_values=None,
                                   strategy="constant")),
    ("vectorization", text_columns_preprocessing),
])
```

```python
categorical_pipeline = Pipeline(steps=[
    ("fill_missing", SimpleImputer(strategy="most_frequent")),
    ("one_hot_encoding", OneHotEncoder(handle_unknown="ignore")),
])
```

```python
numerical_pipeline = Pipeline(steps=[
    ("fill_missing", SimpleImputer(strategy="most_frequent")),
])
```

```python
preprocessing_pipeline = ColumnTransformer(transformers=[
    ("text_columns", text_pipeline, text_columns),
    ("categorical_columns", categorical_pipeline, categorical_columns),
    ("numerical_columns", numerical_pipeline, numerical_columns),
])
```

```python
preprocessing_pipeline.fit(filtered)
```

```python
preprocessed = preprocessing_pipeline.transform(filtered)
```
